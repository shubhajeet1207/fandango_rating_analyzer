Rotten Tomatoes(RT) has two sets of reviews, their critics reviews (ratings published by official critics) and user reviews. 

```python
plt.figure(figsize=(10,4),dpi=150)
sns.scatterplot(data=all_sites,x='RottenTomatoes',y='RottenTomatoes_User')
plt.xlim(0,100)
plt.ylim(0,100)
```

>Quantifying the difference by comparing the critics ratings and the RT User ratings. We will calculate it with RottenTomatoes-RottenTomatoes_User. Note: Rotten_Diff here is Critics - User Score. So values closer to 0 means aggrement between Critics and Users. Larger positive values means critics rated much higher than users. Larger negative values means users rated much higher than critics.

```python
all_sites['Rotten_Diff']  = all_sites['RottenTomatoes'] - all_sites['RottenTomatoes_User']
```

Now comparing the overall mean difference. 
**Since we're dealing with differences that could be negative or positive so, first we'll take the absolute value of all the differences, then take the mean. This would report back on average to absolute difference between the critics rating versus the user rating.**

```python
all_sites['Rotten_Diff'].apply(abs).mean()
```

**Plotting the distribution of the differences between RT Critics Score and RT User Score. There would be negative values in this distribution plot.**

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(data=all_sites,x='Rotten_Diff',kde=True,bins=25)
plt.title("(RT Critics Score) - (RT User Score)");
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.histplot(x=all_sites['Rotten_Diff'].apply(abs),bins=25,kde=True)
plt.title("Abs Difference between RT Critics Score and RT User Score");
```

**Finding out which movies are causing the largest differences. First, we'll show the top 5 movies with the largest *negative* difference between Users and RT critics. Since we calculated the difference as Critics Rating - Users Rating, then large negative values imply the users rated the movie much higher on average than the critics did.**

```python
print("Users Love but Critics Hate")
all_sites.nsmallest(5,'Rotten_Diff')[['FILM','Rotten_Diff']]
```

```python
print("Critics love, but Users Hate")
all_sites.nlargest(5,'Rotten_Diff')[['FILM','Rotten_Diff']]
```

## Fandago Scores vs. All Sites

```python
df = pd.merge(fandango,all_sites,on='FILM',how='inner')
```

```python
df.info()
```

```python
df.head()
```

```python
df['RT_Norm'] = np.round(df['RottenTomatoes']/20,1)
df['RTU_Norm'] =  np.round(df['RottenTomatoes_User']/20,1)
```

```python
df['Meta_Norm'] =  np.round(df['Metacritic']/20,1)
df['Meta_U_Norm'] =  np.round(df['Metacritic_User']/2,1)
```

```python
df['IMDB_Norm'] = np.round(df['IMDB']/2,1)
```

```python
df.head()
```

```python
norm_scores = df[['STARS','RATING','RT_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm']]
```

```python
norm_scores.head()
```

```python
def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
```

```python
fig, ax = plt.subplots(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores,clip=[0,5],shade=True,palette='Set1',ax=ax)
move_legend(ax, "upper left")
```

**We can see clearly that `Fandango` has an uneven distribution. We can also see that `RT critics` have the most uniform distribution. Let's directly compare these two.** 

```python
fig, ax = plt.subplots(figsize=(15,6),dpi=150)
sns.kdeplot(data=norm_scores[['RT_Norm','STARS']],clip=[0,5],shade=True,palette='Set1',ax=ax)
move_legend(ax, "upper left")
```

**Creating a histplot comparing all normalized scores.**

```python
plt.subplots(figsize=(15,6),dpi=150)
sns.histplot(norm_scores,bins=50)
```

```python
sns.clustermap(norm_scores,cmap='magma',col_cluster=False)
```
